## Question 1

```python
def linear_merge(list1, list2):
    list3 = []
    i = j = 0

    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            list3.append(list1[i])
            i += 1
        else:
            list3.append(list2[j])
            j += 1

    while i < len(list1):
        list3.append(list1[i])
        i += 1

    while j < len(list2):
        list3.append(list2[j])
        j += 1

    return list3


list1 = [1, 3, 5, 7, 9]
list2 = [0, 2, 4, 6, 8]
print(f"Input:\nList1: {list1}\nList2: {list2}")
print(f"Output:\nList3: {linear_merge(list1, list2)}")
```

## Question 2

```python
def initialise_dictionary(employees, defaults):
    return {emp: defaults.copy() for emp in employees}


employees = ['Kelly', 'Emma', 'John']
defaults = {"designation": 'Application Developer', "salary": 8000}
print(f"Input:\nemployees = {employees}\ndefaults = {defaults}")
print(f"Output:\n{initialise_dictionary(employees, defaults)}")
```

## Question 3

```python
def delete_keys(sampleDict, keys):
    return {k: v for k, v in sampleDict.items() if k not in keys}


sampleDict = {"name": "Kelly", "age": 25, "salary": 8000, "city": "New york"}
keys = ["name", "salary"]
print(f"Input:\nsampleDict = {sampleDict}\nkeys = {keys}")
print(f"Output:\n{delete_keys(sampleDict, keys)}")
```

## Question 4

```python
def set_operations(x, y):
    w = x.union(y)
    z = x.intersection(y)
    return w - z


x = {1, 2, 3, 4, 5, 6}
y = {5, 6, 7, 8, 9, 10}
print(f"Input:\nx = {x}\ny = {y}")
print(f"Output:\n{set_operations(x, y)}")
```

## Question 5

```python
def divide_numbers(p, q):
    try:
        result = p / q
        return result
    except ZeroDivisionError:
        return "Error: Cannot divide by zero."
    except TypeError:
        return "Error: Please provide numerical values."
    except Exception as e:
        return f"An unexpected error occurred: {e}"


p1, q1 = 10, 0
print(f"Input:\nP = {p1}, Q = {q1}")
print(f"Output:\n{divide_numbers(p1, q1)}")
```
