## Task 1: Reverse String

```python
def reverse_string(s):
    left = 0
    right = len(s)-1

    while left < right:
        temp = s[left]
        s[left] = s[right]
        s[right] = temp
        left += 1
        right -= 1


s = ["h", "e", "l", "l", "o"]
print(f"Input: {s}")
reverse_string(s)
print(f"Output: {s}")

s = ["H", "a", "n", "n", "a", "h"]
print(f"\nInput: {s}")
reverse_string(s)
print(f"Output: {s}")
```

## Task 2: Valid Palindrome

```python
def valid_palindrome(s):
    string = ""

    for char in s.lower():
        if "a" <= char <= "z" or "A" <= char <= "Z" or "0" <= char <= "9":
            string += char

    left = 0
    right = len(string) - 1

    while left < right:
        if string[left] != string[right]:
            return False

        left += 1
        right -= 1

    return True


s = "A man, a plan, a canal: Panama"
print(f"Input: {s}")
print(f"Output: {valid_palindrome(s)}")

s = "race a car"
print(f"\nInput: {s}")
print(f"Output: {valid_palindrome(s)}")
```

## Task 3: Sqrt(x) without any built-in methods

```python
def sqrt_manual(x):
    if x < 2:
        return x
    
    i = 0

    while i * i <= x:
        i += 1

    i -= 1
    return i


x = 4
print(f"Input: {x}")
print(f"Output: {sqrt_manual(x)}")

x = 8
print(f"\nInput: {x}")
print(f"Output: {sqrt_manual(x)}")
```
