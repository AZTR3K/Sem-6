## Task 2

```python
def string_from_both_ends(string):
    if len(string) < 3:
        return ""
    else:
        res = string[0:3]
        n = len(string) - 1
        s = n - 2
        res += string[s: n + 1]
        return res


string = "intelligence"
print(f"Input: {string}")
print((f"Output: {string_from_both_ends(string)}"))
string = "or"
print(f"Input: {string}")
print((f"Output: {string_from_both_ends(string)}"))
```

## Task 3

```python
def replace_occurrences(s):
    freq = {}
    res = ""

    for char in s:
        if char.lower() not in freq:
            freq[char.lower()] = 1
        else:
            char = "@"
        res += char

    return res


string = "Ooogle"
print(f"Input: {string}")
print((f"Output: {replace_occurrences(string)}"))
```

## Task 4

```python
def string_jumble(a, b):
    lista = list(a)
    listb = list(b)

    lista[0], listb[0] = listb[0], lista[0]
    lista[1], listb[1] = listb[1], lista[1]

    a = "".join(lista)
    b = "".join(listb)

    return a + " " + b


a = "mix"
b = "pod"
print(f"A: {a}")
print(f"B: {b}")
print(f"Output: {string_jumble(a, b)}")
a = "dog"
b = "dinner"
print(f"\nA: {a}")
print(f"B: {b}")
print(f"Output: {string_jumble(a, b)}")
```

## Task 5

```python
def match_letters(strings):
    count = 0
    for s in strings:
        if len(s) < 2:
            continue

        if s[0] == s[-1]:
            count += 1

    return count


strings = ["aba", "ab", "aa", "a", "aabaa"]
print(f"Input: {strings}")
print((f"Output: {match_letters(strings)}"))
```

## Task 8

```python
def detect_palindrome(s):
    low = 0
    high = len(s) - 1

    while low <= high:
        if s[low] != s[high]:
            return False
        low += 1
        high -= 1

    return True


s = "aba"
print(f"Input: {s}")
print(f"Output: {detect_palindrome(s)}")
```

## Task 9

```python
def list_flattening(integers):
    res = []

    for nums in integers:
        for num in nums:
            res.append(num)

    return res


integers = [[1, 2, 3], [4, 5]]
print(f"Input: {integers}")
print(f"Output: {list_flattening(integers)}")
```

## Task 10

```python
def list_intersection(numsa, numsb):
    freq = {}
    res = []

    for num in numsa:
        if num in numsb:
            if num not in freq:
                res.append(num)
                freq[num] = 1

    for num in numsb:
        if num in numsa:
            if num not in freq:
                res.append(num)
                freq[num] = 1

    return res


numsa = [1, 2, 3, 3, 4]
numsb = [2, 3, 3, 4, 5, 6, 7]
print(f"Input: {numsa} and {numsb}")
print(f"Output: {list_intersection(numsa, numsb)}")
```

## Task 12

```python
def binary_search(nums, target):
    low = 0
    high = len(nums) - 1

    index = -1

    while low <= high:
        mid = (low + high) // 2

        if nums[mid] == target:
            index = mid
            break
        elif nums[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return index


nums = [1, 2, 3, 4, 5, 6, 7, 8, 9]
target = 8
print(f"Input: {nums}, target: {target}")
print(f"Output: {binary_search(nums, target)}")
```

## Task 13

```python
def merge(arr, low, mid, high):
    l1 = arr[low: mid + 1]
    l2 = arr[mid + 1: high + 1]

    i = 0
    j = 0
    k = low

    while i < len(l1) and j < len(l2):
        if l1[i] <= l2[j]:
            arr[k] = l1[i]
            i += 1
        else:
            arr[k] = l2[j]
            j += 1
        k += 1

    while i < len(l1):
        arr[k] = l1[i]
        i += 1
        k += 1

    while j < len(l2):
        arr[k] = l2[j]
        j += 1
        k += 1


def merge_sort(arr, low, high):
    if low < high:
        mid = (low + high) // 2

        merge_sort(arr, low, mid)
        merge_sort(arr, mid + 1, high)
        merge(arr, low, mid, high)


nums = [38, 27, 43, 3, 9, 82, 10]
print(f"Input: {nums}")
merge_sort(nums, 0, len(nums) - 1)
print(f"Output: {nums}")
```

## Task 14

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)


nums = [38, 27, 43, 3, 9, 82, 10]
print(f"Input: {nums}")
print(f"Output: {quicksort(nums)}")
```

## Task 15

```python
def freq_counter(nums):
    freq = {}

    for num in nums:
        if num not in freq:
            freq[num] = 1
        else:
            freq[num] += 1

    return freq


nums = [1, 2, 2, 3, 4, 2]
print(f"Input: {nums}")
print(f"Output: {freq_counter(nums)}")
```

## Task 16

```python
def count_distinct_elements(nums):
    freq = {}
    count = 0

    for num in nums:
        if num not in freq:
            freq[num] = 1
            count += 1

    return count


nums = [1, 2, 2, 3, 4, 2]
print(f"Input: {nums}")
print(f"Output: {count_distinct_elements(nums)}")
```

```python
def union_and_intersection(set1, set2):
    s3 = set1.union(set2)
    s4 = set1.intersection(set2)
    return (s3, s4)


s1 = {1, 2, 3, 4, 5}
s2 = {2, 3, 4, 5, 6, 7, 8, 9}
print(f"Input: {s1} and {s2}")
print(f"Output: {union_and_intersection(s1, s2)}")
```

## Task 18

```python
def unique_dictionary_values(dictionary):
    freq = {}
    res = []

    for v in dictionary.values():
        if v not in freq:
            freq[v] = 1
            res.append(v)

    return res


dictionary = {1: 2, 2: 3, 3: 4, 4: 5, 5: 5}
print(f"Input: {dictionary}")
print(f"Output: {unique_dictionary_values(dictionary)}")
```

## Task 19

```python
def most_common_values(d):
    if not d:
        return []

    value_counts = {}
    for val in d.values():
        value_counts[val] = value_counts.get(val, 0) + 1

    max_freq = max(value_counts.values())

    most_common_vals = [
        v for v, count in value_counts.items() if count == max_freq]

    return [k for k, v in d.items() if v in most_common_vals]


dictionary = {1: 2, 2: 3, 3: 4, 4: 5, 5: 5}
print(f"Input: {dictionary}")
print(f"Output: {most_common_values(dictionary)}")
```

## Task 20

```python
def combine_dicts(d1, d2):
    new_dict = d2.copy()
    new_dict.update(d1)
    return new_dict


d1 = {1: 2, 2: 3, 3: 4}
d2 = {4: 5, 5: 6}
print(f"Input: {d1} and {d2}")
print(f"Output: {combine_dicts(d1, d2)}")
```

```python
def filter_dicts(dict_list, target_key):
    return [d for d in dict_list if target_key in d]


dict_list = [
    {"name": "Alice", "age": 25},
    {"name": "Bob", "major": "CS"},
    {"name": "Charlie", "age": 30, "major": "Data Science"},
    {"city": "Lahore"}
]

target_key = "name"

print(f"Input: {dict_list}\nTarget: {target_key}")
print(f"Output: {filter_dicts(dict_list, target_key)}")
```

## Task 22

```python
def find_duplicates(items):
    seen = set()
    duplicates = set()

    for item in items:
        if item in seen:
            duplicates.add(item)
        else:
            seen.add(item)

    return list(duplicates)


items = ["python", "java", "c++"]
print(f"Input: {items}")
print(f"Output: {find_duplicates(items)}")
```
