## Task 5: Dataset Handling using Iris Dataset

```python
# TODO:
# 1. Load Iris dataset using pandas
# 2. Display first 5 rows, shape, columns
# 3. Compute mean of numerical attributes
# 4. Find maximum sepal length
# 5. Check for missing values
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
iris_df = pd.read_csv(url, names=columns)
print(f"First five rows:\n{iris_df.head()}")
print("Shape of the DataFrame:", iris_df.shape)
print("Column names:", iris_df.columns)

numeric_cols = iris_df.select_dtypes(include=[np.number])
for col in numeric_cols:
    mean = iris_df[col].mean()
    print(f"Mean of {col}: {mean}")

max = iris_df["sepal_length"].max()
print(f"Max sepal length: {max}")

print(f"Missing values:\n{iris_df.isnull().sum()}")
```

## Task 6: Statistical Measures

```python
# TODO:
# Using a numeric column from Iris dataset:
# Calculate mean, median, mode, range
col = iris_df["sepal_length"]
mean = col.mean()
median = col.median()
mode = col.mode()[0]
value_range = col.max() - col.min()

print("Sepal Length col statistics:")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Range: {value_range}")
```

## Task 7: Data Visualization using Matplotlib

```python
# TODO:
# 1. Line plot: Sepal Length vs Sepal Width
# 2. Scatter plot: Petal Length vs Petal Width
# 3. Histogram of Sepal Length
# Add title, labels, and grid
```

```python
# Line plot
plt.plot(iris_df['sepal_length'], iris_df['sepal_width'],
         marker='o', linestyle='-', color='b', alpha=0.6)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.grid(True)
```

```python
# Scatter plot
plt.scatter(iris_df['petal_length'],
            iris_df['petal_width'], color='red', alpha=0.6)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.grid(True)
```

```python
# Histogram
plt.hist(iris_df['sepal_length'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.grid(True)
```

## Bonus Task: Box Plot Visualization

```python
# TODO:
# Create a box plot for Petal Length
# Identify outliers visually
plt.boxplot(iris_df['petal_length'], patch_artist=True,
            boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot of Petal Length')
plt.ylabel('Length (cm)')
plt.xticks([1], ['Petal Length'])
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
```
